# Podela na ulazne i ciljne atribute

Ciljna promenljiva kojom ćemo meriti stepen promene načina života tokom pandemije COVID-19 ćemo konstruisati na osnovu atributa: 'prod_inc', 'sleep_bal', 'new_skill', 'fam_connect', 'relaxed' i 'self_time', pa je potrebno ove kolone i izbaciti iz ulaznog skupa podataka.

Zašto baš ovi atributi?
1. Matrica korelacije:
Na osnovu analize matrice korelacije zaključili smo da su ovi atributi imali najviše uticaja na ostale atribute koji se odnose na promenu načina života, što ukazuje na njihov potencijalni značaj.

2. Direktan uticaj na psihološko stanje:
Svaki od odabranih atributa direktno se odnosi na aspekte koji mogu značajno uticati na psihološko stanje pojedinaca. Na primer, produktivnost može biti povezano sa osećajem uspeha, opuštenost često povezujemo sa nivoom stresa, režim spavanja ima duboku vezu sa kvalitetom sna i smanjenjem umora, porodica često ima dosta veliki uticaj na nas same,...

Na osnovu  ocena koje opisuju stepen promene konkretnog aspekta, svaku osobu ćemo klasifikovati u jednu od tri kategorija:
* Negativna promena (-1): o < -0.25
* Neutralna promena  (0): -0.25 <= o <= 0.25
* Pozitivna promena  (1): o > 0.25

gde je o srednja vrednost pojedinačnih ocena.

```python
# računanje srednje vrednosti ocena atributa za svakog pojedinca
data['avg_score'] = data[["prod_inc", "sleep_bal", "new_skill", "fam_connect", "relaxed", "self_time"]].mean(axis = 1)

# definisanje pragova za klasifikaciju
positive_threshold = 0.25
negative_threshold = -0.25

# kreiranje ciljne promenljive
data['lifestyle_change'] = pd.cut(data['avg_score'], 
                                  bins = [-float('inf'), negative_threshold, positive_threshold, float('inf')],
                                  labels = [-1, 0, 1])

# brisanje kolona
data.drop(['avg_score'], axis = 1, inplace = True)
data.drop(["prod_inc", "sleep_bal", "new_skill", "fam_connect", "relaxed", "self_time"], axis = 1, inplace = True)

data.head()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# kreiranje countplot sa sortiranim kategorijama 
sns.countplot(x = data["lifestyle_change"], order = sorted(set(data["lifestyle_change"])), ax = axs[0], palette = 'pastel')
axs[0].set_xlabel('Categories')
axs[0].set_ylabel('Count')

# kreiranje pie chart-a 
color = [sns.color_palette('pastel')[1], sns.color_palette('pastel')[2], sns.color_palette('pastel')[0]]
gender_map = data['lifestyle_change'].value_counts()
axs[1].pie(gender_map, labels = gender_map.index, autopct = '%.1f%%', colors = color, startangle = 140)

fig.suptitle("Lifestyle change")
plt.show()
```

S obzirom na ove vrednosti, možemo utvrditi da podaci nisu savršeno balansirani, ali razlike nisu drastične.
Za sada nećemo primeniti nikakve tehnike balansiranja. Ukoliko primetimo da neravnoteža negativno utiče na performanse modela, tada ćemo primeniti neku tehniku za balansiranje klasa.

## Podela podataka na trening i test skup

### Standardizacija podataka

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train = pd.DataFrame(data = X_train_scaled, columns = X_train.columns)
X_test = pd.DataFrame(data = X_test_scaled, columns = X_test.columns)
```
